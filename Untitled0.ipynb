```python
# Importa a biblioteca NumPy e dá a ela o apelido "np"
# Ela é usada para fazer cálculos com números e criar listas de dados numéricos (arrays)
import numpy as np

# Importa a biblioteca pandas e dá a ela o apelido "pd"
# Usamos o pandas para organizar os dados em tabelas chamadas DataFrames, parecidas com planilhas
import pandas as pd

# Importa a biblioteca matplotlib e dá o apelido "plt"
# Essa biblioteca é usada para criar gráficos que ajudam a visualizar os dados
import matplotlib.pyplot as plt

# Importa o modelo IsolationForest da biblioteca scikit-learn
# Esse modelo é uma ferramenta de inteligência artificial usada para detectar anomalias (valores fora do normal) nos dados
from sklearn.ensemble import IsolationForest
```

```python
np.random.seed(42)
n_normais = 300
n_anomalias = 20
```

```python
# --- Geração dos dados normais

# Aceleração no eixo X com média 0.02 g e desvio padrão 0.005 g
acc_x_normal = np.random.normal(0.02, 0.005, n_normais)

# Aceleração no eixo Y com média 0.015 g e desvio padrão 0.004 g
acc_y_normal = np.random.normal(0.015, 0.004, n_normais)

# Velocidade RMS com média 2.0 mm/s e desvio padrão 0.5 mm/s
vel_rms_normal = np.random.normal(2.0, 0.5, n_normais)
```

```python
# --- Geração dos dados anômalos

# Aceleração no eixo X com valores significativamente mais altos (média 0.08 g)
acc_x_anom = np.random.normal(0.08, 0.01, n_anomalias)

# Aceleração no eixo Y com valores mais altos (média 0.07 g)
acc_y_anom = np.random.normal(0.07, 0.015, n_anomalias)

# Velocidade RMS com valores mais altos (média 6.0 mm/s)
vel_rms_anom = np.random.normal(6.0, 1.0, n_anomalias)
```

```python
# --- Combinação dos dados

# Isso permite que o modelo de IA analise todo o conjunto de dados como um só.
acc_x = np.concatenate([acc_x_normal, acc_x_anom])       # Une os dados de aceleração no eixo X, juntando os valores normais e anômalos em um único array.
acc_y = np.concatenate([acc_y_normal, acc_y_anom])       # Une os dados de aceleração no eixo Y (valores normais + anômalos).
vel_rms = np.concatenate([vel_rms_normal, vel_rms_anom]) # Une os dados de velocidade RMS (valores normais + anômalos) em mm/s

# Criação de um DataFrame com os dados combinados
df = pd.DataFrame({
    'Acc_X_g': acc_x,         # Aceleração no eixo X (em g)
    'Acc_Y_g': acc_y,         # Aceleração no eixo Y (em g)
    'Vel_mm_s': vel_rms       # Velocidade RMS (em mm/s)
})
```

```python
modelo = IsolationForest(contamination=0.05, random_state=42)
```

```python
modelo.fit(df)
df['Status'] = modelo.predict(df)
```

```python
# Solicita ao usuário a inserção dos dados de vibração
print("\nDigite os dados de vibração do motor:")

# Coleta os valores de aceleração no eixo X, eixo Y e velocidade RMS
acc_x_user = float(input("Aceleração eixo X (g): "))
acc_y_user = float(input("Aceleração eixo Y (g): "))
vel_user = float(input("Velocidade RMS (mm/s): "))

# Cria um novo DataFrame com os dados informados pelo usuário
entrada = pd.DataFrame({
    'Acc_X_g': [acc_x_user],
    'Acc_Y_g': [acc_y_user],
    'Vel_mm_s': [vel_user]
})
```

```python
# Realiza a predição usando o modelo treinado
status = modelo.predict(entrada)[0]  # Retorna -1 (anomalia) ou 1 (normal)
```

```python
# --- Exibição do diagnóstico

# Mostra o resultado da análise baseada nos dados informados
print("\nDiagnóstico do motor:")
if status == -1:
    print("Vibração fora do padrão! Possível falha detectada.")
else:
    print("Vibração dentro do esperado.")
```

```python
# Esse valor é armazenado em uma nova coluna do DataFrame chamada 'Acc_Total', que será usada no gráfico.
df['Acc_Total'] = np.sqrt(df['Acc_X_g']**2 + df['Acc_Y_g']**2)

# Aqui, fazemos o mesmo cálculo, mas para os valores de aceleração digitados pelo usuário.
acc_user_total = np.sqrt(acc_x_user**2 + acc_y_user**2)

# Criação do gráfico
plt.figure(figsize=(8,6))  # Define o tamanho da figura

# Define as cores com base no status (verde para normal, vermelho para anomalia)
cores = df['Status'].map({1: 'green', -1: 'red'}) # substitui os números por cores

# Plota os pontos históricos (dados simulados)
plt.scatter(df['Vel_mm_s'], df['Acc_Total'], c=cores, alpha=0.6, label='Histórico')

# Plota o ponto do usuário com cor azul e contorno preto
plt.scatter(vel_user, acc_user_total, color='blue', s=100, edgecolor='black', label='Entrada do usuário')

# Configurações do gráfico
plt.xlabel('Velocidade RMS (mm/s)')         # Rótulo do eixo X
plt.ylabel('Aceleração total (g)')          # Rótulo do eixo Y
plt.title("Detecção de Anomalias na Vibração do Motor")  # Título do gráfico
plt.legend()                                # Adiciona a legenda
plt.grid(True)                              # Ativa o grid
plt.tight_layout()                          # Ajusta layout para evitar cortes

# Exibe o gráfico
plt.show()
```
